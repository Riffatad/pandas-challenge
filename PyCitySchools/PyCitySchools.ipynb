```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
unique_schools_count = school_data_complete['school_name'].nunique()
print(f"Total number of unique schools: {unique_schools_count}")

```

```python
# Calculate the total number of students
total_students = school_data_complete.shape[0]
print(f"Total number of students: {total_students}")


```

```python
# Calculate the total budget
# Get unique budget values
unique_budgets = school_data_complete['budget'].unique()

# Calculate the total of unique budget values
total_budget = unique_budgets.sum()

# Print the total budget formatted as currency
print(f"Total Budget: ${total_budget:,.2f}")
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete['math_score'].mean()
print(f"Average_Math_Score: {average_math_score:,.2f}")

```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete['reading_score'].mean()
print(f"Average_Reading_Score: {average_reading_score:,.2f}")
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(total_students) * 100
print(f"Passing Math Percentage: {passing_math_percentage :,.2f}")
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage =  passing_reading_count / float(total_students) * 100
print(f"Passing Reading Percentage: {passing_reading_percentage :,.2f}")
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(total_students) * 100
print(f"overall_passing_rate: {overall_passing_rate :,.2f}")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
dict = {
    "Total Schools": [unique_schools_count],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Reading Score": [average_reading_score],
    "Average Math Score": [average_math_score],
    "Passing Reading Percentage": [passing_reading_percentage],
    "Passing Math Percentage": [passing_math_percentage],
    "Overall Passing Rate": [overall_passing_rate]
    
}
district_summary = pd.DataFrame(dict)

# Format the DataFrame
district_summary["Total Schools"] = district_summary["Total Schools"].map("{:,.0f}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,.2f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Passing Reading Percentage"] = district_summary["Passing Reading Percentage"].map("{:.2f}".format)
district_summary["Passing Math Percentage"] = district_summary["Passing Math Percentage"].map("{:.2f}".format)
district_summary["Overall Passing Rate"] = district_summary["Overall Passing Rate"].map("{:.2f}".format)

# Display the DataFrame
display(district_summary)
```

## School Summary

```python
# Calculate the total student count per school from school_data
student_count_per_school = school_data_complete.groupby("school_name")["Student ID"].count()

# Print the result
#print(student_count_per_school)
```

```python
# Calculate the total school budget per school
per_school_budget = school_data_complete.groupby('school_name')['budget'].unique().astype(int)

# Calculate per capita spending
per_school_capita = per_school_budget/student_count_per_school

# Print the results
#print(per_school_budget)
#print(per_school_capita)
```

```python
# Calculate the average test scores per school from school_data_complete
per_school_math = school_data_complete.groupby('school_name')['math_score'].mean()

per_school_reading = school_data_complete.groupby('school_name')['reading_score'].mean()

# Print the results
#print(per_school_math)
#print(per_school_reading)
```

```python
# Calculate average math and reading scores for each school
per_school_math = school_data_complete.groupby('school_name')['math_score'].mean()
per_school_reading = school_data_complete.groupby('school_name')['reading_score'].mean()
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete
# Filter the DataFrame for students with reading scores of 70 or higher
students_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70]
students_passing_math= school_data_complete[school_data_complete['math_score'] >= 70]

# Calculate the number of students per school with reading scores of 70 or higher
school_students_passing_reading = students_passing_reading.groupby('school_name').size()
school_students_passing_math = students_passing_math.groupby('school_name').size()


# Print the results
#print("Number of students per school with reading scores of 70 or higher:")
#print(school_students_passing_reading)
```

```python
# Calculate the number of students per school that passed both math and reading with scores of 70 or higher
# Filter the DataFrame for students who passed both math and reading with scores of 70 or higher

students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & 
    (school_data_complete["math_score"] >= 70)
]

# Calculate the number of students per school who passed both subjects
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby("school_name").size()

(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)

school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()

# Print the results
#print("Number of students per school who passed both math and reading with scores of 70 or higher:")
#print(school_students_passing_math_and_reading)
```

```python
# Use the provided code to calculate the passing rates
# Calculate the passing rates
per_school_passing_math = (school_students_passing_math / student_count_per_school) * 100
per_school_passing_reading = (school_students_passing_reading / student_count_per_school) * 100
overall_passing_rate = (school_students_passing_math_and_reading / student_count_per_school) * 100


# Print the results
#display("Passing Rate for Math per School:")
#display(per_school_passing_math)

#display("Passing Rate for Reading per School:")
#display(per_school_passing_reading)

#display("Overall Passing Rate per School (both Math and Reading):")
#display(overall_passing_rate)
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "Math Passing Rate": per_school_passing_math,
    "Reading Passing Rate": per_school_passing_reading,
    "Overall Passing Rate": overall_passing_rate,
    "School Type" : school_types
    
})

# Format the budget columns
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
#per_school_summary.index.name = None
display(per_school_summary)
```

## Scores by School Spending

```python
# Define the bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

```

```python
# Create a copy of the school summary for later aggregations
school_spending_df = per_school_summary.copy()

# Use `pd.cut` on the per_school_capita Series from earlier to categorize per student spending based on the bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita,spending_bins, labels = labels)

# Convert Spending Ranges (Per Student) to a string
school_spending_df["Spending Ranges (Per Student)"] = school_spending_df["Spending Ranges (Per Student)"].astype(str)

display(school_spending_df)
```

```python
#  Calculate averages for the desired columns

spending_reading_scores = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["Average Reading Score"].mean()
spending_math_scores = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["Average Math Score"].mean()
spending_math_passing_rate = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["Math Passing Rate"].mean()
spending_reading_passing_rate = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["Reading Passing Rate"].mean()
overall_passing_rate = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["Overall Passing Rate"].mean()

# Display results
#display(spending_reading_scores)
#display(spending_math_scores)
#print(spending_math_passing_rate)
#print(spending_reading_passing_rate)
#print(overall_passing_rate)
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Reading Scores" :spending_reading_scores,
    "Average Math Scores": spending_math_scores,
    "Math Passing Rate": spending_math_passing_rate,
    "Reading Passing Rate": spending_reading_passing_rate,
     "Overall Passing":overall_passing_rate
})
                     

# Display results
display(spending_summary)
```
